```python
from binoculars import Binoculars
import pyarrow.parquet as pq
import pandas as pd
import random
from google.colab import data_table
data_table.enable_dataframe_formatter()
```

```python
# Assumes you have downloaded the data from https://huggingface.co/datasets/artem9k/ai-text-detection-pile to your Google Drive

filenames = [
    "/content/drive/MyDrive/ai-text-detection-pile/data/train-00000-of-00007-bc5952582e004d67.parquet",
    "/content/drive/MyDrive/ai-text-detection-pile/data/train-00001-of-00007-71c80017bc45f30d.parquet",
    "/content/drive/MyDrive/ai-text-detection-pile/data/train-00002-of-00007-ee2d43f396e78fbc.parquet",
    "/content/drive/MyDrive/ai-text-detection-pile/data/train-00003-of-00007-529931154b42b51d.parquet",
    "/content/drive/MyDrive/ai-text-detection-pile/data/train-00004-of-00007-b269dc49374a2c0b.parquet",
    "/content/drive/MyDrive/ai-text-detection-pile/data/train-00005-of-00007-3dce5e05ddbad789.parquet",
    "/content/drive/MyDrive/ai-text-detection-pile/data/train-00006-of-00007-3d8a471ba0cf1c8d.parquet",
]
```

```python
bino = Binoculars()
```

```python
def run_eval(sample_rate, length_requirement=None):
    true_positives = []
    false_positives = []
    true_negatives = []
    false_negatives = []
    error_count = 0

    for filename in filenames:
        parquet_file = pq.ParquetFile(filename)
        print(f"Loaded {filename}")

        for batch in parquet_file.iter_batches(batch_size=1000):
            batch_pandas = batch.to_pandas()

            for index, row in batch_pandas.iterrows():
                if random.random() < sample_rate:
                    if length_requirement and len(row.text) <= length_requirement:
                        continue
                    try:
                        if length_requirement:
                            predicted_ai = (bino.predict(row.text[:length_requirement]) == "Most likely AI-generated")
                        else:
                            predicted_ai = (bino.predict(row.text) == "Most likely AI-generated")
                    except Exception as e:
                        print(f"Error predicting for text: {row.text[:length_requirement] if length_requirement else row.text}, Error: {e}")
                        error_count += 1
                        continue  # Skip to the next iteration if there's an error as happens with a very small number of texts which are empty

                    actually_ai = (row.source == "ai")

                    if predicted_ai and actually_ai:
                        true_positives.append(row.text)
                    elif predicted_ai and not actually_ai:
                        false_positives.append(row.text)
                    elif not predicted_ai and actually_ai:
                        false_negatives.append(row.text)
                    elif not predicted_ai and not actually_ai:
                        true_negatives.append(row.text)

    return {
        'true_positives': true_positives,
        'false_positives': false_positives,
        'true_negatives': true_negatives,
        'false_negatives': false_negatives,
        'error_count': error_count
    }

```

```python
res = run_eval(1.0 / 50)

true_positives_count = len(res['true_positives'])
false_positives_count = len(res['false_positives'])
true_negatives_count = len(res['true_negatives'])
false_negatives_count = len(res['false_negatives'])
error_count = res['error_count']

TPR = true_positives_count / (true_positives_count + false_negatives_count) if (true_positives_count + false_negatives_count) > 0 else 0
FPR = false_positives_count / (false_positives_count + true_negatives_count) if (false_positives_count + true_negatives_count) > 0 else 0
Error_rate = error_count / (false_positives_count + true_negatives_count + true_positives_count + false_negatives_count) if (false_positives_count + true_negatives_count + true_positives_count + false_negatives_count) > 0 else 0

print(f"TPR is {TPR}")
print(f"FPR is {FPR}")
print(f"Error rate is {Error_rate}")
```
